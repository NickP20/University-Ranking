```python
import pandas as pd
```

```python
data_df = pd.read_csv("Data/data.csv")
state_df = data_df[['state_full','name']].groupby('state_full').count().rename(columns={'name':'Total Count'})
state_df = state_df.sort_values('Total Count',ascending = False)
state_df=state_df.reset_index()
state_df
```

```python
state_df.to_json("Data/pieplot.json", orient = "records")
```
