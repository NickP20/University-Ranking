## 2022 Best Colleges in America - A Web Scraping from NICHE

### Data Source
- https://www.niche.com/colleges/search/best-colleges/

<span style="color:blue">*The Best Colleges ranking is based on rigorous analysis of academic, admissions, financial, and student life data from the U.S. Department of Education along with millions of reviews from students and alumni. The ranking compares more than 1,000 top colleges and universities in the U.S. This year's rankings have reduced the weight of ACT/SAT scores to reflect a general de-emphasis on test scores in the college admissions process.*</span>

### Methodology
#### Factors considered
- Academics Grade - 40.0%
- Value Grade - 27.5%
- Professors Grade - 7.5%
- Campus Grade - 5.0%
- Diversity Grade - 5.0%
- Student Life Grade - 5.0%
- Student Surveys on Overalll Experience - 5.0%
- Local Area Grade - 2.5%
- Safety Grade - 2.5% 

<strong>Information of the top 800 colleges is exported in .csv file.</strong>

#### import dependencies

```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time
```

#### scraping

```python
college_info = []
# scrape pages 1-32, the top 800 colleges, ranking ends at 800 
for p in range(1,33):
    #Setup splinter
    executable_path = {'executable_path': ChromeDriverManager().install()}
    browser = Browser('chrome',**executable_path, headless = False)
    url = f'https://www.niche.com/colleges/search/best-colleges/?page={p}'
    browser.visit(url)
    html = browser.html
    soup = BeautifulSoup(html,'html.parser')
    colleges = soup.find_all('div',class_='card__inner')
    print(f"Page{p}------------------------------------------------------------------------")
    
    for college in colleges:
        try:
            #-----college name
            name = college.h2.text
            #-----college rating
            rating = college.find('div',class_='search-result-badge').text
        
            college_tag = college.find('ul',class_='search-result-tagline').find_all('li')
            #-----college location by city and state
            location = college_tag[1].text.strip()  
        
            #grade, acceptance rate, net price, SAT range
            college_fact = college.find('ul',class_='search-result-fact-list').find_all('li')
            #-----Overall Niche Grade
            nonBreakSpace = u'\xa0'
            grade = college_fact[0].find('div',class_='niche__grade').text.replace(nonBreakSpace, ' ')
            #-----acceptance rate
            acceptance_rate = college_fact[1].find('span').text
            #-----Net Price
            net_price = college_fact[2].find('span').text
            #-----SAT range
            SAT_range = college_fact[3].find('span').text
            info = [name,rating,location,grade,acceptance_rate,net_price,SAT_range]
            college_info.append(info)
                
            print(name)
            print(rating)
            print(location)
            print(grade)
            print(acceptance_rate)
            print(net_price)
            print(SAT_range)
            print('-'*40)
            
        except AttributeError as e:
            pass
    
    browser.quit()
    time.sleep(1)
```

```python
len(college_info)
```

#### assemble data into a pandas dataframe

```python
college_df = pd.DataFrame(college_info,columns = ['name','rating','location','niche_grade','acceptance_rate (%)','net_price ($)','SAT_range'])
college_df.head()
```

#### clean up the data

```python
# There are missing values in net price and SAT_range columns
college_df = college_df.replace('—', np.NaN)
```

```python
college_df['rating'] = range(1,1+len(college_df))
college_df['city'] = college_df['location'].str.split(',').str[0]
college_df['state'] = college_df['location'].str.split(',').str[1]
college_df['acceptance_rate (%)'] = college_df['acceptance_rate (%)'].str.replace('%','')
college_df['niche_grade'] = college_df['niche_grade'].str.split(' ').str[1]
college_df['net_price ($)'] = college_df['net_price ($)'].replace('[\$,]', '', regex=True).astype(float).astype('Int64')
college_df['SAT_low'] = college_df['SAT_range'].str.split('-').str[0].astype('float').astype('Int64')
college_df['SAT_high'] = college_df['SAT_range'].str.split('-').str[1].astype('float').astype('Int64')
college_df['SAT_average'] = (college_df['SAT_low'] + college_df['SAT_high'])/2
college_df.head(10)
```

#### export csv

```python
college_df.to_csv('top800_best_colleges_in_America_2022.csv')
```
